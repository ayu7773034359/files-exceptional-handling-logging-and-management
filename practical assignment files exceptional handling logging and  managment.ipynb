```python
filename = "example.txt"

try:
    # Try to open the file for reading
    with open(filename, "r") as file:
        # Read each line in the file and print it
        for line in file:
            print(line.strip())
except FileNotFoundError:
    # Handle the error if the file does not exist
    print(f"The file '{filename}' does not exist.")
```

 4.Write a Python script that reads from one file and writes its content to another file

```python
source_file = "source.txt"
destination_file = "destination.txt"

try:
    # Open the source file for reading
    with open(source_file, "r") as src:
        # Open the destination file for writing
        with open(destination_file, "w") as dest:
            # Read each line from the source file and write it to the destination file
            for line in src:
                dest.write(line)
    print(f"Contents of '{source_file}' have been copied to '{destination_file}'.")

except FileNotFoundError:
    print(f"The file '{source_file}' does not exist.")
except Exception as e:
    print(f"An error occurred: {e}")
```

 8. Write a program to handle a file opening error using exception handling?

```python
def read_file(filename):
    try:
        # Attempt to open the file
        with open(filename, 'r') as file:
            content = file.read()
            print("File content:")
            print(content)
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except IsADirectoryError:
        print(f"Error: Expected file but found a directory: '{filename}'.")
    except PermissionError:
        print(f"Error: Permission denied when trying to open '{filename}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    # Specify the filename to read
    filename = "example.txt"  # Change this to a file you want to test
    read_file(filename)
```

 9.How can you read a file line by line and store its content in a list in Python?

```python
def read_file_to_list(filename):
    content_list = []  # Initialize an empty list to store lines

    try:
        with open(filename, 'r') as file:
            for line in file:
                content_list.append(line.strip())  # Adding each line to the list, stripping newline characters
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except IsADirectoryError:
        print(f"Error: Expected file but found a directory: '{filename}'.")
    except PermissionError:
        print(f"Error: Permission denied when trying to open '{filename}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

    return content_list

if __name__ == "__main__":
    filename = "example.txt"  # Change this to the file you want to read
    lines = read_file_to_list(filename)
    print("File content as list:")
    print(lines)
```

 10.How can you append data to an existing file in Python?

```python
def append_to_file(filename, data):
    try:
        # Open the file in append mode
        with open(filename, 'a') as file:
            file.write(data + '\n')  # Write the data followed by a newline
            print("Data appended successfully.")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except PermissionError:
        print(f"Error: Permission denied when trying to write to '{filename}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    filename = "example.txt"  # Change this to your file
    data_to_append = "This is the new line of text."  # Data you want to append
    append_to_file(filename, data_to_append)
```

 12.Write a program that demonstrates using multiple except blocks to handle different types of exceptions?

```python
def demonstrate_exceptions():
    numbers = [10, 0, 5]
    invalid_index = 4
    result = None

    # Example of handling multiple exceptions
    try:
        # Division operation which can raise ZeroDivisionError
        result = numbers[0] / numbers[1]
        print(f"Result of division: {result}")

        # Index operation which can raise IndexError
        value = numbers[invalid_index]
        print(f"Value at index {invalid_index}: {value}")

        # Attempting to add incompatible types which can raise TypeError
        result = numbers[0] + "Hello"
        print(f"Result of addition: {result}")

    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
    except IndexError:
        print(f"Error: Index {invalid_index} is out of range for the list.")
    except TypeError:
        print("Error: Incompatible types for the operation.")

if __name__ == "__main__":
    demonstrate_exceptions()
```

 14.Write a program that uses the logging module to log both informational and error messages

```python
import logging

# Configure the logging
logging.basicConfig(
    level=logging.DEBUG,  # You can change this to INFO to log only info and above
    format='%(asctime)s - %(levelname)s - %(message)s',
    filename='app.log',  # Log to a file named 'app.log'
    filemode='a'  # Append mode
)

def perform_division(a, b):
    logging.info(f'Attempting to divide {a} by {b}')
    try:
        result = a / b
        logging.info(f'Result of division: {result}')
        return result
    except ZeroDivisionError as e:
        logging.error(f'Error occurred: {e}')
        return None

def main():
    logging.info('Program started')

    # Example of logging informational messages and handling errors
    perform_division(10, 2)  # Should succeed
    perform_division(10, 0)  # Should raise and log a ZeroDivisionError

    logging.info('Program finished')

if __name__ == "__main__":
    main()
```

 15.Write a Python program that prints the content of a file and handles the case when the file is empty?

```python
def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()

            # Check if the content is empty
            if not content:
                print("The file is empty.")
            else:
                print("File content:")
                print(content)

    except FileNotFoundError:
        print(f"The file '{file_path}' does not exist.")
    except IOError as e:
        print(f"An error occurred while reading the file: {e}")

def main():
    file_path = 'example.txt'  # Specify the file name here
    read_file(file_path)

if __name__ == "__main__":
    main()

```

 21.Write a Python program that reads a file and prints the number of occurrences of a specific word

```python
def count_word_in_file(file_path, target_word):
    try:
        # Open the file using a context manager
        with open(file_path, 'r', encoding='utf-8') as file:
            # Initialize a counter for the word occurrences
            word_count = 0

            # Read the file line by line
            for line in file:
                # Split the line into words and count occurrences of the target word
                words = line.split()
                word_count += words.count(target_word)

            print(f"The word '{target_word}' occurs {word_count} times in the file '{file_path}'.")

    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except IOError as e:
        print(f"Error occurred while reading the file: {e}")

if __name__ == "__main__":
    # Specify the path to your file and the word to count
    file_path = 'example.txt'  # Change this to your file's path
    target_word = 'Python'      # Change this to the word you want to count

    count_word_in_file(file_path, target_word)
```

 22.How can you check if a file is empty before attempting to read its contents?

```python
import os

def is_file_empty(file_path):
    try:
        return os.path.exists(file_path) and os.stat(file_path).st_size == 0
    except FileNotFoundError:
        return False  # File doesn't exist
    except IOError as e:
        print(f"Error occurred while accessing the file: {e}")
        return False

file_path = 'example.txt'  # Change this to your file's path

if is_file_empty(file_path):
    print(f"The file '{file_path}' is empty.")
else:
    print(f"The file '{file_path}' is not empty.")
```

 23.Write a Python program that writes to a log file when an error occurs during file handling.

```python
import os
import logging

# Setup logging configuration
logging.basicConfig(filename='file_error.log', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def read_file(file_path):
    try:
        # Attempt to read the file
        with open(file_path, 'r', encoding='utf-8') as file:
            contents = file.read()
            print(contents)

    except FileNotFoundError as e:
        logging.error(f"FileNotFoundError: {e}")
        print(f"Error: The file '{file_path}' was not found.")

    except IOError as e:
        logging.error(f"IOError: {e}")
        print(f"Error occurred while reading the file: {e}")

if __name__ == "__main__":
    # Specify the path to your file
    file_path = 'example.txt'  # Change this to your target file path

    read_file(file_path)
```
